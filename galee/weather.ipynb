```python
# 12년도부터 22년도까지의 일 평균 기온그래프
 

heat_12 = pd.read_csv('data\weather_data_mean_month_2012.csv', encoding='cp949')
heat_13 = pd.read_csv('data\weather_data_mean_month_2013.csv', encoding='cp949')
heat_14 = pd.read_csv('data\weather_data_mean_month_2014.csv', encoding='cp949')
heat_15 = pd.read_csv('data\weather_data_mean_month_2015.csv', encoding='cp949')
heat_16 = pd.read_csv('data\weather_data_mean_month_2016.csv', encoding='cp949')
heat_17 = pd.read_csv('data\weather_data_mean_month_2017.csv', encoding='cp949')
heat_18 = pd.read_csv('data\weather_data_mean_month_2018.csv', encoding='cp949')
heat_19 = pd.read_csv('data\weather_data_mean_month_2019.csv', encoding='cp949')
heat_20 = pd.read_csv('data\weather_data_mean_month_2020.csv', encoding='cp949')
heat_21 = pd.read_csv('data\weather_data_mean_month_2021.csv', encoding='cp949')
heat_22 = pd.read_csv('data\weather_data_mean_month_2022.csv', encoding='cp949')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 8]

df =  pd.concat([heat_12, heat_13,heat_14, heat_15,heat_16, heat_17,heat_18, heat_19, heat_20, heat_21, heat_22]).reset_index(drop=True)


# Define the colors for the line plot
# colors = ['blue', 'orange', 'green', 'red', 'black','gray', 'pink','yellow', 'sky', 'purple']

# Create a line plot for each year
for i, year in enumerate(df['year'].unique()):
    year_data = df[df['year'] == year]
    plt.plot(year_data['month'], year_data['기온(°C)'], label=str(year))

# Set the x-axis and y-axis labels
plt.xlabel('Month')
plt.ylabel('Power Demand (MW)')

# Set the x-axis tick labels (months)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(1, 13), months)

# Add a legend to the plot
plt.legend()

# Set the title of the plot
plt.title('temp(°C) by Month and Year')

# Show the plot
plt.show()
```

```python
heat_18 = pd.read_csv('data\sunpower_mean_month_2018.csv', encoding='cp949')
heat_19 = pd.read_csv('data\sunpower_mean_month_2019.csv', encoding='cp949')
heat_20 = pd.read_csv('data\sunpower_mean_month_2020.csv', encoding='cp949')
heat_21 = pd.read_csv('data\sunpower_mean_month_2021.csv', encoding='cp949')

weather = pd.read_csv('월별_전국_평균날씨.csv', encoding='cp949')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 8]

df_heat = pd.concat([heat_18, heat_19, heat_20, heat_21]).reset_index(drop=True)
df_weather = pd.concat([weather]).reset_index(drop=True)

df_merged = pd.merge(df_heat, df_weather, on=['year', 'month'])

corr = df_merged['기온(°C)'].corr(df_merged['발전량 '])
# Drop unnecessary columns
df_merged.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y'], inplace=True)

# Rename columns for better readability
df_merged.rename(columns={'power generation': '발전량 ', 'current temp(°C)': '기온(°C)'}, inplace=True)

plt.scatter(df_merged['기온(°C)'], df_merged['발전량 '])

# Add axis labels and title
plt.xlabel('current temp(°C)')
plt.ylabel('Generation(kWh)')
plt.title(f'Correlation between temp(°C) and Generation: {corr}')
# print(corr)
# Show the plot
plt.show()
```
