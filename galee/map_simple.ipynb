```python
import pandas as pd
```

```python
data = pd.read_csv('태양광 발전량_test.csv', encoding='cp949')
data.head(5)
```

```python
import folium,json
```

```python
jsonfile = open('TL_SCCO_CTPRVN.json','r',encoding='utf8').read()
jsondata = json.loads(jsonfile)
jsondata
```

```python
m = folium.Map(location=[36.6719633,127.9961565],zoom_start=5)
cho= folium.Choropleth(geo_data=jsondata, data=data,columns=['지역','발전량'],
                  key_on='feature.properties.CTP_KOR_NM', legend_name='전국 태양광',
                  bins=10,fill_color='RdPu').add_to(m)

cho.geojson.add_child(
    folium.features.GeoJsonTooltip(['CTP_KOR_NM'],labels=False)
)
m
```
