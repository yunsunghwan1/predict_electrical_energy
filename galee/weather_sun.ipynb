```python
import pandas as pd
import folium,json
import ipywidgets as widgets
from ipywidgets import interact
import datetime
```

```python
# 태양광 생산량을 18년도부터 21년도까지의 데이터를 월별로 히트맵으로 표시


heat_18 = pd.read_csv('data\sunpower_mean_month_2018.csv', encoding='cp949')
heat_19 = pd.read_csv('data\sunpower_mean_month_2019.csv', encoding='cp949')
heat_20 = pd.read_csv('data\sunpower_mean_month_2020.csv', encoding='cp949')
heat_21 = pd.read_csv('data\sunpower_mean_month_2021.csv', encoding='cp949')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 8]

df =  pd.concat([heat_18, heat_19, heat_20, heat_21]).reset_index(drop=True)


data = df.pivot(index='year', columns='month', values='발전량 ').to_numpy()

# Define the x-axis tick labels (months)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# Define the y-axis tick labels (years)
years = df['year'].unique()
fig, ax = plt.subplots()
im = ax.imshow(data, cmap='YlGnBu')

ax.set_xticks(np.arange(len(months)))
ax.set_yticks(np.arange(len(years)))
ax.set_xticklabels(months)
ax.set_yticklabels(years)

plt.setp(ax.get_xticklabels(), rotation=45, ha="right", rotation_mode="anchor")

# Add the data values as annotations to the heatmap
# for i in range(len(years)):
#     for j in range(len(months)):
#         text = ax.text(j, i, data[i, j], ha="center", va="center", color="w")

# Add a colorbar to the heatmap
cbar = ax.figure.colorbar(im, ax=ax)

# Set the title of the plot
ax.set_title("Power Generation (in kWh)")

# Show the plot
plt.show()
```

```python
# 태양광 생산량을 18년도부터 21년도까지의 데이터를 년도별로 점그래프로 표시

heat_18 = pd.read_csv('data\sunpower_mean_month_2018.csv', encoding='cp949')
heat_19 = pd.read_csv('data\sunpower_mean_month_2019.csv', encoding='cp949')
heat_20 = pd.read_csv('data\sunpower_mean_month_2020.csv', encoding='cp949')
heat_21 = pd.read_csv('data\sunpower_mean_month_2021.csv', encoding='cp949')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 8]

df = pd.concat([heat_18, heat_19, heat_20, heat_21]).reset_index(drop=True)

# Define the colors for the scatter plot
colors = ['blue', 'orange', 'green', 'red']

# Create a scatter plot for each year
for i, year in enumerate(df['year'].unique()):
    year_data = df[df['year'] == year]
    plt.scatter(year_data['month'], year_data['발전량 '], c=colors[i], label=str(year))

# Set the x-axis and y-axis labels
plt.xlabel('Month')
plt.ylabel('Power Generation (in kWh)')

# Set the x-axis tick labels (months)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(1, 13), months)

# Add a legend to the plot
plt.legend()

# Set the title of the plot
plt.title('Power Generation by Month and Year')

# Show the plot
plt.show()
```

```python
# 태양광 생산량을 18년도부터 21년도까지의 데이터를 년도별로 꺾은선그래프로 표시

heat_18 = pd.read_csv('data\sunpower_mean_month_2018.csv', encoding='cp949')
heat_19 = pd.read_csv('data\sunpower_mean_month_2019.csv', encoding='cp949')
heat_20 = pd.read_csv('data\sunpower_mean_month_2020.csv', encoding='cp949')
heat_21 = pd.read_csv('data\sunpower_mean_month_2021.csv', encoding='cp949')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 8]

df = pd.concat([heat_18, heat_19, heat_20, heat_21]).reset_index(drop=True)

# Define the colors for the line plot
colors = ['blue', 'orange', 'green', 'red']

# Create a line plot for each year
for i, year in enumerate(df['year'].unique()):
    year_data = df[df['year'] == year]
    plt.plot(year_data['month'], year_data['발전량 '], c=colors[i], label=str(year))

# Set the x-axis and y-axis labels
plt.xlabel('Month')
plt.ylabel('Power Generation (in kWh)')

# Set the x-axis tick labels (months)
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
plt.xticks(np.arange(1, 13), months)

# Add a legend to the plot
plt.legend()

# Set the title of the plot
plt.title('Power Generation by Month and Year')

# Show the plot
plt.show()
```

```python
# 태양광 생산량을 18년도부터 21년도까지의 데이터를 년도별로 꺾은선그래프로 표시


heat_18 = pd.read_csv('data\sunpower_mean_month_2018.csv', encoding='cp949')
heat_19 = pd.read_csv('data\sunpower_mean_month_2019.csv', encoding='cp949')
heat_20 = pd.read_csv('data\sunpower_mean_month_2020.csv', encoding='cp949')
heat_21 = pd.read_csv('data\sunpower_mean_month_2021.csv', encoding='cp949')
demand_18 = pd.read_csv('data\power_demand_mean_month_2018.csv', encoding='cp949')
demand_19 = pd.read_csv('data\power_demand_mean_month_2019.csv', encoding='cp949')
demand_20 = pd.read_csv('data\power_demand_mean_month_2020.csv', encoding='cp949')
demand_21 = pd.read_csv('data\power_demand_mean_month_2021.csv', encoding='cp949')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 8]

df_heat = pd.concat([heat_18, heat_19, heat_20, heat_21]).reset_index(drop=True)
df_demand = pd.concat([demand_18, demand_19, demand_20, demand_21]).reset_index(drop=True)

df_merged = pd.merge(df_heat, df_demand, on=['year', 'month'])

corr = df_merged['현재수요(MW)'].corr(df_merged['발전량 '])
# Drop unnecessary columns
df_merged.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y'], inplace=True)

# Rename columns for better readability
df_merged.rename(columns={'power generation': '발전량 ', 'current demand (MW)': 'demand'}, inplace=True)

plt.scatter(df_merged['현재수요(MW)'], df_merged['발전량 '])

# Add axis labels and title
plt.xlabel('Demand (MW)')
plt.ylabel('Generation (kWh)')
plt.title(f'Correlation between Power Demand and Generation: {corr}')

# Show the plot
plt.show()
```

```python
# 기온과 전력수요량의 상관계수 그래프 18~21년
# 점은 월 평균을, 선은 상관계수를 나타냄

heat_18 = pd.read_csv('data\sunpower_mean_month_2018.csv', encoding='cp949')
heat_19 = pd.read_csv('data\sunpower_mean_month_2019.csv', encoding='cp949')
heat_20 = pd.read_csv('data\sunpower_mean_month_2020.csv', encoding='cp949')
heat_21 = pd.read_csv('data\sunpower_mean_month_2021.csv', encoding='cp949')
demand_18 = pd.read_csv('data\power_demand_mean_month_2018.csv', encoding='cp949')
demand_19 = pd.read_csv('data\power_demand_mean_month_2019.csv', encoding='cp949')
demand_20 = pd.read_csv('data\power_demand_mean_month_2020.csv', encoding='cp949')
demand_21 = pd.read_csv('data\power_demand_mean_month_2021.csv', encoding='cp949')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 8]

df_heat = pd.concat([heat_18, heat_19, heat_20, heat_21]).reset_index(drop=True)
df_demand = pd.concat([demand_18, demand_19, demand_20, demand_21]).reset_index(drop=True)

df_merged = pd.merge(df_heat, df_demand, on=['year', 'month'])


# Calculate the correlation coefficient
corr = df_merged['현재수요(MW)'].corr(df_merged['발전량 '])

# Drop unnecessary columns
df_merged.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y'], inplace=True)

# Rename columns for better readability
df_merged.rename(columns={'power generation': '발전량 ', 'current temp(°C)': '현재수요(MW)'}, inplace=True)

# Create a scatter plot of the data
plt.scatter(df_merged['현재수요(MW)'], df_merged['발전량 '])

# Draw the correlation coefficient line
plt.plot(np.unique(df_merged['현재수요(MW)']), np.poly1d(np.polyfit(df_merged['현재수요(MW)'], df_merged['발전량 '], 1))(np.unique(df_merged['현재수요(MW)'])), color='red')

# Add axis labels and title
plt.xlabel('Demand (MW)')
plt.ylabel('Generation(kWh)')
plt.title(f'Correlation between Demand (MW) and Generation: {corr}')

# Show the plot
plt.show()
```

```python
# 기온에 따른 전력 수요량 박스플롯 18~21년
# 일단위로 표시, 기온를 2°C단위로 끊어서 표현

heat_18 = pd.read_csv('data2\일별_태양광 발전량.csv', encoding='cp949')

weather = pd.read_csv('data2\일별_전국_평균날씨.csv', encoding='cp949')


import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 8]

df_heat = pd.concat([heat_18]).reset_index(drop=True)
df_weather = pd.concat([weather]).reset_index(drop=True)

df_trans = []

for i, item in enumerate(df_weather['기온(°C)']):
    item = int(item / 2) * 2
    df_trans.append(item)

df_weather['기온'] = df_trans

# df_weather.head(2)

df_merged = pd.merge(df_heat, df_weather, on=['year', 'month'])
df_merged.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y'], inplace=True)
df_merged.rename(columns={'power generation': '발전량', 'current temp(°C)': '기온'}, inplace=True)


sns.boxplot(x=df_merged['기온'], y=df_merged['발전량'])

# Add axis labels and title
plt.xlabel('Temperature(°C)')
plt.ylabel('Generation(kWh)')
plt.title(f'Boxplot of Generation by Temperature')

plt.show()
```

```python
# 풍속에 따른 전력 수요량 박스플롯 18~21년
# 일단위로 표시, 기온를 0.5단위로 끊어서 표현

heat_18 = pd.read_csv('data2\일별_태양광 발전량.csv', encoding='cp949')

weather = pd.read_csv('data2\일별_전국_평균날씨.csv', encoding='cp949')

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.rcParams['figure.figsize'] = [10, 8]

df_heat = pd.concat([heat_18]).reset_index(drop=True)
df_weather = pd.concat([weather]).reset_index(drop=True)

df_trans = []

for i, item in enumerate(df_weather['풍속(m/s)']):
    item = int(item / 0.5) * 0.5
    df_trans.append(item)

df_weather['풍속'] = df_trans

# df_weather.head(2)

df_merged = pd.merge(df_heat, df_weather, on=['year', 'month'])
df_merged.drop(columns=['Unnamed: 0_x', 'Unnamed: 0_y'], inplace=True)
df_merged.rename(columns={'power generation': '발전량', 'current temp(°C)': '풍속'}, inplace=True)


sns.boxplot(x=df_merged['풍속'], y=df_merged['발전량'])

# Add axis labels and title
plt.xlabel('current Wind Speed (m/s)')
plt.ylabel('Generation(kWh)')
plt.title(f'Boxplot of Wind Speed (m/s) by Temperature')

plt.show()
```
