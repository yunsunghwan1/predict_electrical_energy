```python
# 시간 변수를 생성하는 함수 정의
def lag_function(df, var_name, past):
    # reset_index // drop=True 기존에 있던 인덱스를 칼럼으로 저장하지 않겠다. 
    # inplace = True 수정한 내용을 df 데이터 프레임에 저장한다. 
    df.reset_index(drop=True, inplace = True)
    
    #시간변수 생성
    # df의 데이터프레임에 lag+past+_var_name의 칼럼을 생성한 후 그 칼럼에 nan값을 저장한다. 
    df['lag'+str(past)+'_'+var_name] = np.nan
    df['lag'+str(past)+'_'+'AB'] = np.nan
    
    # col 첫번 쨰 값에 AB를 넣는다. 코드 실행이 끝난다. 두번쨰로 col var_name을 넣어서 코드를 실행한다. 
    for col in ['AB', var_name]:
        # df의 데이터 프레임 수 많큼 동작한다. +1을 하는 이유는 range가 0에서 시작하기 때문이다. 
        for i in range(0,(max(df.index)+1)):
            # 데이터 프레임의 값이 batter_name같은 사람을 찾은 후 year- past하여 past많큼 과거의 값으로 저장한다. 
            val = df.loc[(df['batter_name'] == df['batter_name'][i]) & 
                         (df['year'] == df['year'][i] - past), col]
            # 과거 기록이 결측치가 아니라면 값을 넣기/ 즉 값이 있다면 데이터에 저장 .
            if(len(val) != 0):
                df.loc[i, 'lag' + str(past) + '_' + col] = val.iloc[0]

    #30타수 미만 결측치 처리
    df.loc[df['lag' + str(past) + '_' + 'AB'] < 30, 
           'lag' + str(past) + '_' + var_name] = np.nan
    df.drop('lag' + str(past) + '_' + 'AB', axis = 1, inplace = True)


# 요약본

# 이 코드 블록은 데이터 프레임의 각 숫자 열에 대해 지연된 변수를 생성합니다. 데이터 프레임의 각 행을 반복하고 해당 행의 'batter_name' 및 'year' 값을 확인합니다. 
# 그런 다음 이 정보를 사용하여 현재 행 이전의 '지난' 년 수인 행에 대해 동일한 열에서 지정된 열('col')의 값을 찾습니다. 
# 값이 발견되면 접두사 'lag' + 'past' + '_' + 'col'을 사용하여 현재 행의 새 열에 저장됩니다.
# 지난 해의 타석 수가 30개 미만이면 'var_name'에 해당하는 지연 변수가 NaN으로 설정됩니다. 마지막으로 'lag' + 'past' + '_' + 'AB' 열이 데이터 프레임에서 삭제됩니다.



    return df

numeric_cols = list(regular_season_df.select_dtypes(include=numerics).drop(
    ['batter_id','year','OPS','SLG'], axis =1).columns)

# 이 코드는 'batter_id', 'year', 'OPS' 및 'SLG' 열을 제외하고 일반_season_df 데이터 프레임에서 숫자 유형인 열 이름 목록을 만듭니다.
# select_dtypes() 메서드의 'include=numerics' 매개변수는 숫자 데이터 유형(정수, 부동 소수점 등)을 가진 열을 포함합니다.
# .drop() 메서드는 'drop' 목록에 지정된 이름의 컬럼을 삭제하고, 'axis=1' 매개변수는 컬럼 축임을 지정합니다.
# 마지막으로 .columns 속성은 결과 데이터 프레임의 열 이름 목록을 반환합니다.
# 열 이름의 결과 목록은 변수 'numeric_cols'에 저장됩니다.
```
