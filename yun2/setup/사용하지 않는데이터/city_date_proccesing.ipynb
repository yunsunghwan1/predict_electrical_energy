```python
# 필요 라이브러리 로드
from matplotlib import font_manager, rc
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import platform

#그래프에서 마이너스 기호가 표시되게 하는 설정입니다.
plt.rcParams['axes.unicode_minus'] == False
```

```python

path1 = 'data/광주광역시_가구평균_월사용량/가구 평균 월별 전력사용량_2021.01~06.xls'
path2 = 'data/광주광역시_가구평균_월사용량/가구 평균 월별 전력사용량_2021.07~12.xls'
path3 = 'data/광주광역시_가구평균_월사용량/가구 평균 월별 전력사용량_2022.01~06.xls'
path4 = 'data/광주광역시_가구평균_월사용량/가구 평균 월별 전력사용량_2022.07~12.xls'
power_household_average_2021_01 = pd.read_excel(path1)
power_household_average_2021_07 = pd.read_excel(path2)
power_household_average_2022_01 = pd.read_excel(path3)
power_household_average_2022_07 = pd.read_excel(path4)
```

```python
power_household_average_2022_07.head(5)
```

```python
power_household_average_2022_07.drop([0,1],axis=0)
```

```python
# [0,1]을 하여 인데그 0~1의 값을 제거 인덱스 2의 값은 0,2 해도 제거가 되지 않아서 따로 제거
power_household_average_2022_07.drop([0,1],axis=0, inplace=True)
power_household_average_2022_01.drop([0,1],axis=0, inplace=True)
power_household_average_2021_07.drop([0,1],axis=0, inplace=True)
power_household_average_2021_01.drop([0,1],axis=0, inplace=True)
```

```python
power_household_average_2021_01.iloc[0:8]
```

```python
power_household_average_2021_01_entire = power_household_average_2021_01.iloc[1:7]

power_household_average_2021_07_entire = power_household_average_2021_07.iloc[1:7]
power_household_average_2022_01_entire = power_household_average_2022_01.iloc[1:7]
power_household_average_2022_07_entire = power_household_average_2022_07.iloc[1:7]
```

```python
power_household_average_2022_07_entire
```

```python
power_household_entire = pd.concat([power_household_average_2022_07_entire,
                                    power_household_average_2022_01_entire,
                                    power_household_average_2021_07_entire,
                                    power_household_average_2021_01_entire])
```

```python
power_household_entire.sort_values('가구 평균 월별 전력사용량 ( 20230321 현재 )',inplace=True)
```

```python
power_household_entire.drop(columns='Unnamed: 2', inplace=True)
```

```python
power_household_entire.rename(columns={'가구 평균 월별 전력사용량 ( 20230321 현재 )':'년도',
                                                       'Unnamed: 1': "시도",
                                                       'Unnamed: 3': "대상가구수(호)",
                                                       'Unnamed: 4': "가구당 평균 전력 사용량(kWh)",
                                                       'Unnamed: 5': "가구당 평균 전기요금(원)"}, inplace=True)
```

```python
power_household_entire.reset_index(drop=True, inplace=True)
```
