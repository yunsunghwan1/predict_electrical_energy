```python
# 필요 라이브러리 로드
from matplotlib import font_manager, rc
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import platform

#그래프에서 마이너스 기호가 표시되게 하는 설정입니다.
plt.rcParams['axes.unicode_minus'] == False
```

```python

path1 = 'data/광주광역시_용도별_월사용량/계약종별 전력사용량 월별_2021.01~06.xls'
path2 = 'data/광주광역시_용도별_월사용량/계약종별 전력사용량 월별_2021.07~12.xls'
path3 = 'data/광주광역시_용도별_월사용량/계약종별 전력사용량 월별_2022.01~06.xls'
path4 = 'data/광주광역시_용도별_월사용량/계약종별 전력사용량 월별_2022.07~12.xls'
power_jobtype_2021_01 = pd.read_excel(path1)
power_jobtype_2021_07 = pd.read_excel(path2)
power_jobtype_2022_01 = pd.read_excel(path3)
power_jobtype_2022_07 = pd.read_excel(path4)
```

```python
power_jobtype_2021_01.head(12)
```

```python
power_jobtype_2021_01.drop(labels=range(0, 48),axis=0,inplace=True)
power_jobtype_2021_07.drop(labels=range(0, 48),axis=0, inplace=True)
power_jobtype_2022_01.drop(labels=range(0, 48),axis=0, inplace=True)
power_jobtype_2022_07.drop(labels=range(0, 48),axis=0, inplace=True)
```

```python
power_jobtype_2021_01[:-1]
```

```python
power_jobtype_2021_01 = power_jobtype_2021_01.iloc[:-1]
power_jobtype_2021_07 = power_jobtype_2021_07.iloc[:-1]
power_jobtype_2022_01 = power_jobtype_2022_01.iloc[:-1]
power_jobtype_2022_07 = power_jobtype_2022_07.iloc[:-1]
```

```python
power_entire = pd.concat([power_jobtype_2021_01,
                                    power_jobtype_2021_07,
                                    power_jobtype_2022_01,
                                    power_jobtype_2022_07])
```

```python
power_entire.reset_index(drop=True, inplace=True)
```

```python
power_entire
```

```python
power_entire.rename(columns={'계약종별 전력사용량 월별( 20230321 현재 )':'년도',
                                                       'Unnamed: 1': "시도",
                                                       'Unnamed: 1': "시군구",
                                                       'Unnamed: 3': "계약구분",
                                                       'Unnamed: 4': "고객수",
                                                       'Unnamed: 5': "사용량(kWh)",
                                                       'Unnamed: 6': "전기요금(원)",
                                                       'Unnamed: 7': "평균판매단가(원/kWh)"}, inplace=True)
```

```python
power_entire.sort_values('년도',inplace=True)
```
